```python
import pandas as pd
```

```python
data= pd.read_csv('gen.csv')
```

```python
data
```

```python
data = data.sort_values(by=['Company', 'Year'])
```

```python
for col in data.columns:
  if data[col].dtype == 'object':
    data[col] = data[col].str.replace(',', '')
```

```python
data=data.apply(pd.to_numeric, errors='ignore')
```

```python
data['Revenue YoY Change (%)'] = data.groupby('Company')['Total revenue'].pct_change(fill_method=None) * 100
```

```python
data['Net Income Growth (%)'] = data.groupby(['Company'])['Net income'].pct_change(fill_method=None) * 100
```

```python
data['Total Assets YoY Change (%)'] = data.groupby('Company')['Total assests'].pct_change(fill_method=None) * 100
```

```python
data['Total Liabilities YoY Change (%)'] = data.groupby('Company')['Total liabilities'].pct_change(fill_method=None) * 100
```

```python
data['Net Cash from Operations YoY Change (%)'] = data.groupby('Company')['Net cash from operations'].pct_change(fill_method=None) * 100
```

**Tesla**: Experienced significant revenue and income growth in the past but is now facing a slowdown and challenges in maintaining profitability, with a notable reduction in assets and liabilities.

**Apple**: Demonstrates resilience with stable assets and a recent recovery in income and cash flow, though revenue trends are not available for analysis.

Overall, all three companies show signs of adjusting to market conditions, with some facing challenges in maintaining previous growth levels, while others are stabilizing or recovering.
